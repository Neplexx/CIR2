## <font color = green> Mission 1 : horloge détraquée </font>

On cherche un "pas" qu'on va nommer k de façons qu'avec un nombre x d'étapes, on atteigne l'entier 1, il faut donc que x soit l'inverse de k car nous travaillons dans le monoide (Z/nZ,x) ayant pour neutre 0, l'équation à résoudre est donc l'opération de k par x modulo n qui est égal au neutre : k*x = 1 [n]

Dans le cas de 7 avec n = 60 (horloge) on a : 7 x 43 = 301 => 301 = (5*60) + 1 donc le reste est bien de 1 et 7 x 43 = 1 [60]

En pratique, on veut savoir si notre entier est inversible dans le monoide(Z/nZ,x), pour ce faire on calcul d'abord le pgcd entre notre entier k et n , si le résultat est un entier p différent de 1 : alors on observera sur l'aiguille suivra le cycle des multiples de p sans jamais passer par 1

Une fois cette première étape validée, on établit la relation de Bézout, dans le cas de 7 : 7x + 60 = 1 

puis on applique l'algorithme d'Euclide ainsi que la rémontée : 

60 = 7x8 + 4

7 = 4x1 + 3

4 = 3x1 + 1

Et donc la rémontée : 

1=4−(3x1)

1=4−(7−4x1)x1=2x4−7

1=2x(60−7x8)−7

1=2x60−16x7−7

1=2x60−17x7

On passe ensuite au modulo, le 60 devient la classe 0 : notre inverse est donc -17, on applique le cycle de 60 pour obtenir notre "réprésentant" de la classe de -17 qui est finalement 43

```python
n = 60
R = Integers(n)
possibles = [k for k in range(n) if gcd(k, n) == 1]

k = R(7) #exemple avec 7
y = R(6) # et avec 6
try:
    x = k^-1
    print(f"Le 'pas' 7 est possible avec un total de {x} étapes")
except:
    print("Non inversible")

try:
    p = y^-1
except:
    print("Non inversible")
```
