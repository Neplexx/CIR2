## <font color = green> Mission 1 : horloge détraquée </font>

On cherche un "pas" qu'on va nommer k de façons qu'avec un nombre x d'étapes, on atteigne l'entier 1, il faut donc que x soit l'inverse de k car nous travaillons dans le monoide (Z/nZ,x) ayant pour neutre 0, l'équation à résoudre est donc l'opération de k par x modulo n qui est égal au neutre : k*x = 1 [n]

Dans le cas de 7 avec n = 60 (horloge) on a : 7 x 43 = 301 => 301 = (5*60) + 1 donc le reste est bien de 1 et 7 x 43 = 1 [60]

En pratique, on veut savoir si notre entier est inversible dans le monoide(Z/nZ,x), pour ce faire on calcul d'abord le pgcd entre notre entier k et n , si le résultat est un entier p différent de 1 : alors on observera sur l'aiguille suivra le cycle des multiples de p sans jamais passer par 1

Une fois cette première étape validée, on établit la relation de Bézout, dans le cas de 7 : 7x + 60 = 1 

puis on applique l'algorithme d'Euclide ainsi que la rémontée : 

60 = 7x8 + 4

7 = 4x1 + 3

4 = 3x1 + 1

Et donc la rémontée : 

1=4−(3x1)

1=4−(7−4x1)x1=2x4−7

1=2x(60−7x8)−7

1=2x60−16x7−7

1=2x60−17x7

On passe ensuite au modulo, le 60 devient la classe 0 : notre inverse est donc -17, on applique le cycle de 60 pour obtenir notre "réprésentant" de la classe de -17 qui est finalement 43

```python
n = 60
R = Integers(n)
possibles = [k for k in range(n) if gcd(k, n) == 1]

k = R(7) #exemple avec 7
y = R(6) # et avec 6
try:
    x = k^-1
    print(f"Le 'pas' 7 est possible avec un total de {x} étapes")
except:
    print("Non inversible")

try:
    p = y^-1
except:
    print("Non inversible")
```

## <font color = green> Mission 2 : génération pseudo-aléatoire de nombres </font>

```python
n = 21247
x1 = 1936
x2 = 12586
x3 = 8850


Zn = Zmod(n)


x1_Zn = Zn(x1)
x2_Zn = Zn(x2)
x3_Zn = Zn(x3)


diff = x2_Zn - x1_Zn
print(f"Vérification de l'inversibilité de (x2 - x1) = {diff}")
print(f"PGCD({diff}, {n}) = {gcd(diff, n)}")




a = (x3_Zn - x2_Zn) / diff
print(f"\nMultiplicateur a = {a}")



b = x2_Zn - a * x1_Zn
print(f"Incrément b = {b}")


verification = a * x2_Zn + b
print(f"\nVérification: a*x2 + b = {verification}")
print(f"Valeur attendue de x3: {x3_Zn}")
print(f"Succès: {verification == x3_Zn}")



x0 = (x1_Zn - b) / a
print(f"\nGraine x0 = {x0}")


verification_x1 = a * x0 + b
print(f"Vérification: a*x0 + b = {verification_x1}")
print(f"Valeur attendue de x1: {x1_Zn}")
print(f"Succès: {verification_x1 == x1_Zn}")
```

## <font color = green> Mission 3 : la géométrie des tables de multiplication </font>

```python
# tracer une table de multiplication géométrique de a modulo m
def trace_table(a, m):
    points = [(cos(2*pi*k/m), sin(2*pi*k/m)) for k in range(m)]
    G = Graphics()
    for k in range(m):
        depart = points[k]
        arrivee = points[(a*k) % m]
        G += line([depart, arrivee], color='blue')
    return G

# tracer et afficher pour a de 2 à 6 et m = 70
for a in range(2, 7):
    trace_table(a, 70).show(title=f"a={a}, m=70")

# observation des multiplicateurs particuliers
trace_table(35, 70).show() # a = m/2 : 1 lobe
trace_table(34, 70).show() # a = m/2 - 1
trace_table(36, 70).show() # a = m/2 + 1
trace_table(19, 70).show() # a = m/4 + 2 (jolie fleur)
```

### Explications mathématiques des géométries obtenues

**1. Explication du constat : pourquoi $(a-1)$ lobes ?**
Les "creux" qui séparent les lobes d'une figure se forment là où les cordes sont de longueur nulle, c'est-à-dire lorsqu'un point $k$ est relié à lui-même. Mathématiquement, on cherche les points fixes de la table de multiplication :
$$a \times k \equiv k \pmod m$$
Ce qui se réécrit :
$$(a - 1)k \equiv 0 \pmod m$$
Sur un cercle divisé en $m$ parties, cette équation possède exactement $(a - 1)$ solutions uniformément réparties sur le pourtour. Ces points d'ancrage structurent la courbe (appelée *épicycloïde*) et expliquent la présence systématique de $(a-1)$ lobes.

**2. Étude des multiplicateurs particuliers (avec $m=70$ pair)**
*   **Cas $a = \frac{m}{2}$ (ex: $a=35$) :** 
    Si $k$ est pair ($k=2p$), alors $a \times k = \frac{m}{2} \times 2p = mp \equiv 0 \pmod m$. 
    Si $k$ est impair ($k=2p+1$), alors $a \times k = mp + \frac{m}{2} \equiv \frac{m}{2} \pmod m$.
    Toutes les lignes convergent vers seulement deux points diamétralement opposés ($0$ et $35$), formant un simple faisceau de droites.

*   **Cas $a = \frac{m}{2} - 1$ (ex: $a=34$) :** 
    L'équation devient $k \mapsto (\frac{m}{2} - 1)k = \frac{m}{2}k - k$. Cela combine un décalage d'un demi-cercle (pour les impairs) et une symétrie axiale inverse ($-k$). Cela donne une figure très dense à 33 lobes.

*   **Cas $a = \frac{m}{2} + 1$ (ex: $a=36$) :** 
    L'équation est $k \mapsto (\frac{m}{2} + 1)k = \frac{m}{2}k + k$.
    Si $k$ est pair, on obtient $0 + k = k$ (le point reste sur place).
    Si $k$ est impair, on obtient $\frac{m}{2} + k$ (le point est relié au point diamétralement opposé). La figure trace les diamètres du cercle pour les nombres impairs.

*   **Cas $a = \frac{m}{4} + 2$ (ex: $a=19$) :** 
    La présence du diviseur $4$ ($\frac{m}{4}$) introduit un comportement cyclique qui décale les résultats par quarts de cercle ($\frac{\pi}{2}$). Ce comportement impose géométriquement une symétrie d'ordre 4 à la structure, ce qui sculpte les 4 "pétales" de cette jolie fleur algorithmique.


## <font color = green> Mission 4 : étudier une structure algébrique </font>

```python
import itertools
import random

def comp(f,g):
    return {x: f[g[x]] for x in f}

f = {1:2, 2:3, 3:1}
g = {1:1, 2:1, 3:2}

print(comp(f,g))

def all_func(n):
    valeurs = [1..n]
    fonctions = []
    for prod in itertools.product(valeurs, repeat=n):
        f = {i+1: prod[i] for i in range(n)}
        fonctions.append(f)
    return fonctions

n = 3
F = all_func(n)

print(len(F))
print(n^n)

def associative(n):
    F = all_func(n)
    for f in F:
        for g in F:
            for h in F:
                if comp(f, comp(g,h)) != comp(comp(f,g), h):
                    return False
    return True

%time print(associative(3))
%time print(associative(4))

def bijective(f):
    return len(set(f.values())) == len(f)

def count_sym(n):
    F = all_func(n)
    c = 0
    for f in F:
        if bijective(f):
            c += 1
    return c

print(count_sym(5))
print(factorial(5))

def commute(f,g):
    return comp(f,g) == comp(g,f)

def count_commuting(n):
    F = all_func(n)
    c = 0
    for f in F:
        for g in F:
            if commute(f,g):
                c += 1
    return c

print(count_commuting(2))

def orbite(f,x):
    seen = []
    while x not in seen:
        seen.append(x)
        x = f[x]
    return seen

def moyenne_orbite(n, essais=1000):
    total = 0
    for i in range(essais):
        f = {i: randint(1,n) for i in [1..n]}
        for x in [1..n]:
            total += len(orbite(f,x))
    return total/(essais*n)

print(moyenne_orbite(7))
```
