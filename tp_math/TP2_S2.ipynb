On observe que pour calculer le nombre de compositions possibles d'un entier naturel par d'autres entiers naturels, on fait la formule 2^(n-1), mais pourquoi ? Prenons notre entier n qu'on va représenter par n*1, pour chaque 1 on a le choix entre le relier avec le prochain ou non (d'où le 2 : je relie ou non) et cette représentation n'étant pas une "table ronde" le n ième "1" n'est pas relier au premier "1", laissons donc n-1 possibilités de reliaisons. Nous voilà donc avec notre formule 2^(n-1), 
Voici un exemple avec l'entier 3 : 1-1-1 ; 
je les relies tous (1+1+1) = (3)
je relies les 2 premiers mais pas le dernier (1+1,1) = (2,1)
Je reliers pas le premier mais les deux d'après si (1,1+1) = (1,2)
J'en relie aucun : (1,1,1)

Maintenant, dans le calcul de nombres de partitions d'un entier naturel pas d'autres entiers naturels (1,2) et (2,1) représentent tous les deux {1,2} et ne doivent donc pas etres répétés
Pour expliquer le procédé reprenons l'entier 3 : 
si je veux faire un groupe de 3, aucune unité est exclue : (3)
si je veux faire un groupe de 2, une unité est exclue : (2,1)
